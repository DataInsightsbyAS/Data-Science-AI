```python
# Display value counts for all categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

for col in categorical_columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
```

```python
# Missing values 
print(df.isnull().sum())
```

```python
# Missing Value Percentage
missing_count = df.isnull().sum()
missing_percent = (missing_count / len(df)) * 100

missing_summary = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing Percentage (%)': missing_percent
})

print("Missing Value Summary:")
print(missing_summary[missing_summary['Missing Count'] > 0])
```

```python
# Detect Outliers Using IQR
numerical_cols = df.select_dtypes(include=np.number).columns

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"{col}: {len(outliers)} outliers")
```

```python
### Suspicious Values or data entry errors

# Negative values in numerical columns
negative_count = (df[col] < 0).sum() 
if negative_count > 0: 
    print(f"{col}: {negative_count} negative values")
else:
    print("None of the numerical columns contain negative values")

# Percentage columns should be between 0 and 100
percentage_cols = [col for col in df.columns if 'pct' in col.lower() or 'ratio' in col.lower()]
for col in percentage_cols:
    invalid_count = ((df[col] < 0) | (df[col] > 100)).sum()
    if invalid_count > 0:
        print(f"{col}: {invalid_count} values outside the range [0,100]")
else:
    print("Each column percentage is between 0 & 100")

# Enrollment and teachers should not be zero or negative
if 'total_enrollment' in df.columns:
    print("Schools with zero enrollment:",
          (df['total_enrollment'] <= 0).sum())

if 'total_teachers' in df.columns:
    print("Schools with zero teachers:",
          (df['total_teachers'] <= 0).sum())
```

```python
# Identify Redundant Columns
for col in df.columns:
    if df[col].nunique() == 1:
        print(col)
else:
    print("No redundant column")

# Duplicate rows
print("\nDuplicate Rows:", df.duplicated().sum())
```

#### Specific Questions about the dataset

```python
# province having highest enrollment?
highest_enrollment_province = (
    df.groupby('province')['total_enrollment']
    .mean()
    .idxmax()
)
print("Province with the highest enrollment:",
      highest_enrollment_province)
```

```python
# district having highest average pass rate?
highest_pass_district = (
    df.groupby('district')['pass_rate_pct']
    .mean()
    .idxmax()
)

print("District with the highest average pass rate:",
      highest_pass_district)
```

```python
# school type having lowest average dropout rate?
lowest_dropout_school_type = (
    df.groupby('school_type')['dropout_rate_pct']
    .mean()
    .idxmin()
)

print("School type with the lowest average dropout rate:",
      lowest_dropout_school_type)
```

```python
# province having highest average infrastructure score?
highest_infra_province = (
    df.groupby('province')['infrastructure_score']
    .mean()
    .idxmax()
)

print("Province with the highest average infrastructure score:",
      highest_infra_province)
```

```python
# District with the highest number of government schools having internet access
gov_internet = df[
    (df['school_ownership'] == 'Government') &
    (df['has_internet'] == 'Yes')
]['district'].value_counts()

print("District with the highest number of government schools with internet access:")
print(gov_internet.idxmax(), "-", gov_internet.max(), "schools")
```

```python
# Number of private schools that offer scholarships
private_scholarship_count = df[
    (df['school_ownership'] == 'Private') &
    (df['scholarship_available'] == 'Yes')
].shape[0]

print("Number of private schools that offer scholarships:",
      private_scholarship_count)
```

```python
# School type having highest average annual budget?
highest_budget_school_type = (
    df.groupby('school_type')['annual_budget_pkr']
    .mean()
    .idxmax()
)

print("School type with the highest average annual budget:",
      highest_budget_school_type)
```

```python
# Area type with the highest number of poor building conditions
highest_poor_area = (
    df[df['building_condition'] == 'Poor']['area_type']
    .value_counts()
)

print("Area type with the highest number of schools having poor building conditions:",
      highest_poor_area.idxmax())
```
